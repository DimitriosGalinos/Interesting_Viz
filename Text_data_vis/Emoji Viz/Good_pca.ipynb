```python
# -*- coding: utf-8 -*-
"""
Created on Thu Nov 26 17:59:55 2020

@author: mitsel
"""

import json
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

with open('EmojiOnlyModel.json') as json_file:
    data = json.load(json_file)

big_list = []
for idx, sub in enumerate(data, start = 0): 
    big_list.append(list(sub.values())) 

feat_list = [item[0] for item in big_list]
emoji_list = [item[2] for item in big_list]

pca = PCA(n_components=2)
my2D = pca.fit_transform(feat_list)

xxx = my2D[:,0]
yyy = my2D[:,1]
#plt.scatter(x, y)

intermediate_emoj = []

for xx in range(len(emoji_list)):
    if '-' in  emoji_list[xx]:
        g =  emoji_list[xx].split('-')
        emoji_list[xx] = int(g[0], 16)
        intermediate_emoj.append(emoji_list[xx])
        emoji_list[xx]=chr(emoji_list[xx])
    else:
        emoji_list[xx] = int(emoji_list[xx], 16)
        intermediate_emoj.append(emoji_list[xx])
        emoji_list[xx]=chr(emoji_list[xx])
```

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame(np.transpose(np.array([xxx,yyy, emoji_list])),  
                  columns=["1st Principal Component", 
                           "2nd Principal Component",
                           "emojis"])
fig = px.scatter(df, x="1st Principal Component", y="2nd Principal Component",
                text=df["emojis"])
fig.update_layout(title='Emoji visualization using PCA')
fig.show()
```

```python
import plotly.graph_objects as go

fig2 = go.Figure(data=go.Scatter(x=df['1st Principal Component'],
                                y=df["2nd Principal Component"],
                                mode='markers',
                                #marker_color=df["int values"],
                                marker=dict(
                                            size=7,
                                            color=np.add(xxx,yyy), #set color equal to a variable
                                            colorscale='Viridis'
                                        ),
                                text=df['emojis'])) # hover text goes here

fig2.update_layout(title='Emoji visualization using PCA',
                   xaxis_title='1st Principal Component',
                   yaxis_title="2nd Principal Component")
fig2.show()
```
